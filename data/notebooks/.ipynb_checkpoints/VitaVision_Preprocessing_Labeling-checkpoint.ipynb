```python
import pandas as pd

def read_file(file):
    return pd.read_csv(file, sep=None, engine="python")

zinc = read_file("zinc_cleaned.csv")
vit_e = read_file("vitamin_e_cleaned.csv")
vit_a = read_file("vitamin_a_cleaned.csv")
vit_d = read_file("vitamin_d_cleaned.csv")
vit_c = read_file("vitamin_c_cleaned.csv")
magnesium = read_file("magnesium_cleaned.csv")
folate = read_file("folate_cleaned.csv")
ferritin = read_file("ferritin_cleaned.csv")
vit_k = read_file("vitamin_k_cleaned.csv")
b12 = read_file("b12_cleaned.csv")
b6 = read_file("b6_cleaned.csv")
calcium = read_file("calcium_cleaned.csv")
```

```python
for name, df in {
    "zinc": zinc,
    "vit_e": vit_e,
    "vit_a": vit_a,
    "vit_d": vit_d,
    "vit_c": vit_c,
    "magnesium": magnesium,
    "folate": folate,
    "ferritin": ferritin,
    "vit_k": vit_k,
    "b12": b12,
    "b6": b6,
    "calcium": calcium
}.items():
    print(name, df.columns.tolist())
```

```python
calcium.rename(columns={
    'Gender': 'RIAGENDR',
    'Age': 'RIDAGEYR'
}, inplace=True)
```

```python
def classify_standard(value, low, high):
    if pd.isna(value):
        return None
    elif value < low:
        return "Deficient"
    elif value <= high:
        return "Normal"
    else:
        return "Excessive"
```

```python
# Zinc
zinc['Label'] = zinc['Zinc'].apply(lambda x: classify_standard(x, 66, 106))

# Vitamin E
vit_e['Label'] = vit_e['LBXVIE'].apply(lambda x: classify_standard(x, 500, 2000))

# Vitamin A
vit_a['Label'] = vit_a['LBXVIA'].apply(lambda x: classify_standard(x, 28, 86))

# Vitamin D
vit_d['Label'] = vit_d['Vitamin_D'].apply(lambda x: classify_standard(x, 20, 50))

# Vitamin C
vit_c['Label'] = vit_c['LBXVIC'].apply(lambda x: classify_standard(x, 0.4, 2.0))

# Magnesium
magnesium['Label'] = magnesium['Magnesium'].apply(lambda x: classify_standard(x, 1.7, 2.2))

# Folate
folate['Label'] = folate['LBDFOT'].apply(lambda x: classify_standard(x, 3, 20))

# B12
b12['Label'] = b12['LBDB12'].apply(lambda x: classify_standard(x, 200, 900))

# B6
b6['Label'] = b6['LBXPLP'].apply(lambda x: classify_standard(x, 20, 100))

# Vitamin K
vit_k['Label'] = vit_k['Vitamin_K'].apply(lambda x: classify_standard(x, 0.10, 2.20))

# Calcium
calcium['Label'] = calcium['Calcium'].apply(lambda x: classify_standard(x, 8.6, 10.2))
```

```python
def classify_ferritin(row):
    value = row['LBXFER']
    gender = row['RIAGENDR']
    
    if gender == 1:
        return classify_standard(value, 30, 400)
    else:
        return classify_standard(value, 13, 150)

ferritin['Label'] = ferritin.apply(classify_ferritin, axis=1)
```

```python
for name, df in {
    "zinc": zinc,
    "vit_d": vit_d,
    "b12": b12
}.items():
    print("\n", name)
    print(df['Label'].value_counts())
```

```python
all_datasets = {
    "zinc": zinc,
    "vit_e": vit_e,
    "vit_a": vit_a,
    "vit_d": vit_d,
    "vit_c": vit_c,
    "magnesium": magnesium,
    "folate": folate,
    "ferritin": ferritin,
    "vit_k": vit_k,
    "b12": b12,
    "b6": b6,
    "calcium": calcium
}

for name, df in all_datasets.items():
    print("\n", name)
    print(df["Label"].value_counts())
```

```python
zinc.to_csv("zinc_labeled.csv", index=False)
vit_e.to_csv("vitamin_e_labeled.csv", index=False)
vit_a.to_csv("vitamin_a_labeled.csv", index=False)
vit_d.to_csv("vitamin_d_labeled.csv", index=False)
vit_c.to_csv("vitamin_c_labeled.csv", index=False)
magnesium.to_csv("magnesium_labeled.csv", index=False)
folate.to_csv("folate_labeled.csv", index=False)
ferritin.to_csv("ferritin_labeled.csv", index=False)
vit_k.to_csv("vitamin_k_labeled.csv", index=False)
b12.to_csv("b12_labeled.csv", index=False)
b6.to_csv("b6_labeled.csv", index=False)
calcium.to_csv("calcium_labeled.csv", index=False)
```

```python
def prepare_dataset(df, value_col, nutrient_name, age_col="RIDAGEYR", gender_col="RIAGENDR"):
    temp = df[["SEQN", age_col, gender_col, value_col, "Label"]].copy()
    temp.rename(columns={
        age_col: "Age",
        gender_col: "Gender",
        value_col: "Value"
    }, inplace=True)
    temp["Nutrient"] = nutrient_name
    return temp[["SEQN", "Age", "Gender", "Nutrient", "Value", "Label"]]
```

```python
zinc_final = prepare_dataset(zinc, "Zinc", "Zinc")
vit_e_final = prepare_dataset(vit_e, "LBXVIE", "Vitamin E")
vit_a_final = prepare_dataset(vit_a, "LBXVIA", "Vitamin A")
vit_d_final = prepare_dataset(vit_d, "Vitamin_D", "Vitamin D")
vit_c_final = prepare_dataset(vit_c, "LBXVIC", "Vitamin C")
magnesium_final = prepare_dataset(magnesium, "Magnesium", "Magnesium")
folate_final = prepare_dataset(folate, "LBDFOT", "Folate")
ferritin_final = prepare_dataset(ferritin, "LBXFER", "Ferritin")
vit_k_final = prepare_dataset(vit_k, "Vitamin_K", "Vitamin K")
b12_final = prepare_dataset(b12, "LBDB12", "Vitamin B12")
b6_final = prepare_dataset(b6, "LBXPLP", "Vitamin B6")
calcium_final = prepare_dataset(calcium, "Calcium", "Calcium", age_col="RIDAGEYR", gender_col="RIAGENDR")
```

```python
final_dataset = pd.concat([
    zinc_final,
    vit_e_final,
    vit_a_final,
    vit_d_final,
    vit_c_final,
    magnesium_final,
    folate_final,
    ferritin_final,
    vit_k_final,
    b12_final,
    b6_final,
    calcium_final
], ignore_index=True)

final_dataset.head()
```
